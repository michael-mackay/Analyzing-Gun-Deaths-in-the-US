```python
import pandas as pd
import matplotlib as plt

%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
raw_data_count = len(recent_grads.index)
recent_grads = recent_grads.dropna()
```

```python
cleaned_data_count = len(recent_grads.index)
raw_data_count-cleaned_data_count
```

```python
recent_grads.plot(x='Sample_size',y='Median',kind='scatter')
```

There is little correlation found between the number of students in a major and the median salary. The highest salaries are found in majors with few students, but the lowest salaries are found among that group as well. The most popular majors tend to be in the lower-to-middle range of median salary.

```python
recent_grads.plot(x='Sample_size',y='Unemployment_rate',kind='scatter')
```

In the case of unemployment rate, the extremes are found among the least represented majors. The data looks to have somewhat of a normal distribution. It appears that the more popular a major, the less likely that it is represented among the extremes in unemployment rate. 

```python
recent_grads.plot(x='Full_time',y='Median',kind='scatter')
```

The number of full time employees shows a similar relationship to median pay as sample size. There is little correlation.

```python
recent_grads.plot(x='ShareWomen',y='Unemployment_rate',kind='scatter')
```

There is no appreciable correlation between the share of women in a major and the unemployment rate. 

```python
recent_grads.plot(x='Men',y='Median',kind='scatter')
```

```python
recent_grads.plot(x='Women',y='Median',kind='scatter')
```

The number of men or women in a major shows similar correlation to median pay, both resembling the correlation shown with overall sample size.

```python
recent_grads.plot(x='ShareWomen',y='Median',kind='scatter')
```

A slight negative correlation can be observed between the share of women in a major and the median pay.
